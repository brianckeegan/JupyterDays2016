```python
# Standard packages for data visualization
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sb

# numpy lets us handle numeric arrays efficiently
import numpy as np

# pandas handles tabular data
import pandas as pd

# networkx handles network data
import networkx as nx

# To run queries against MediaWiki APIs
import wikitools as wt

# json handles reading and writing JSON data
import json

# To parse links out of HTML markup in Wikipedia pages
from bs4 import BeautifulSoup

# Some other helper functions
from collections import Counter
from operator import itemgetter
```

```python
def extract_query_to_list(json, stuff, key):
    l = []
    pageid = str(json['query']['pages'].keys()[0])
    if stuff in json['query']['pages'][pageid]:
        for item in json['query']['pages'][pageid][stuff]:
            if key:
                l.append(item[key])
            else:
                l.append(item)
    return l

def wikipedia_query(wiki,params,stuff,key):
    """ Perform a query on a 'wiki' endpoint, with a dictionary of parameters, that returns results keyed by 'stuff'
    """
    # Make the query
    results = wt.api.APIRequest(wiki=wiki,data=params).queryGen()
    
    # The results are a generator, so we need to catch the data
    result_list = []
    for result in results:
        result_list += extract_query_to_list(result,stuff,key)
        
    return result_list

def get_revision_df(article):
    revisions_query = {'action': 'query',
                      'redirects': 'True',
                      'prop': 'revisions',
                      'titles': article,
                      'rvprop': 'ids|user|timestamp|user|userid|comment|size',
                      'rvlimit': '500',
                      'rvdir': 'newer'}

    revisions_data = wikipedia_query(enwp,revisions_query,'revisions',None)

    # Extract and convert to DataFrame. Try/except for links to pages that don't exist
    try:
        df = pd.DataFrame(revisions_data)
        
        # Make it clear what's being edited
        df['page'] = article

        # Clean up timestamps
        df['timestamp'] = pd.to_datetime(df['timestamp'],format="%Y-%m-%dT%H:%M:%SZ",unit='s')

        # Clean up anon column. If/else for articles that have all non-anon editors
        if 'anon' in df.columns:
            df = df.replace({'anon':{np.nan:False,u'':True}})
        else:
            df['anon'] = False

        # Sort the data on timestamp and reset the index
        df = df.sort_values('timestamp').reset_index(drop=True)
        df.index.name = 'revision'
        df = df.reset_index()

        # Set the index to a MultiIndex
        #df.set_index(['page','revision'],inplace=True)

        # Compute additional features
        df['date'] = df['timestamp'].apply(lambda x:x.date())
        df['diff'] = df['size'].diff()
        df['latency'] = df['timestamp'].diff().apply(lambda x:round(x/np.timedelta64(1,'s'),-1) + 1)

        # Don't return random other columns
        df = df[[u'anon',u'comment',u'parentid',u'revid',u'size',u'timestamp',
                 u'revision',u'user',u'userid',u'date', u'diff', u'latency','page']]

        return df
        
    except KeyError:
        print u"{0} doesn't exist!".format(article)
        pass
    
enwp = wt.wiki.Wiki(url='https://en.wikipedia.org/w/api.php')
```

```python
merrick_garland_df = get_revision_df('Merrick Garland')
merrick_garland_df.tail()
```

```python
activity_by_day = merrick_garland_df.groupby('date').aggregate({'revid':len})
activity_by_day['cum_revisions'] = activity_by_day['revid'].cumsum()

ax = activity_by_day.plot(lw=5,secondary_y=['cum_revisions'])
#patches, labels = ax.get_legend_handles_labels()
#ax.legend(patches, labels, loc='center left',bbox_to_anchor=(1,.5))

ax.set_ylim((0,50))
ax.set_ylabel('Daily revisions',fontsize=15)
ax.set_xlabel('Time',fontsize=15)
```

```python
activity_by_day = merrick_garland_df.groupby('date').aggregate({'revid':len,
                                                       'diff':np.median,
                                                       'latency':np.median})

# Compute a 7-day rolling average to remove spikiness, plot
activity_by_day['latency'].rolling(window=7).median().plot()
#plt.yscale('symlog')
plt.xlabel('Time',fontsize=15)
plt.ylabel('Latency time (seconds)',fontsize=15)
```

```python
# Compute a 7-day rolling average to remove spikiness, plot
activity_by_day['diff'].rolling(window=7).mean().plot()
#plt.yscale('symlog')
plt.xlabel('Time',fontsize=15)
plt.ylabel('Median changes (bytes)',fontsize=15)
```

Plot the media bytes changes and number of revisions made by date over the time window between Justice Scalia's death on February 13 and the most recent data (on March 18).

```python
_df = activity_by_day.ix[pd.date_range('2016-02-13','2016-03-18'),['diff','revid']]
_df.rename(columns={'diff':'median bytes changed','revid':'number of revisions'},inplace=True)

ax = _df.fillna(0).plot(lw=3)
#ax.axvline(pd.to_datetime('2016-03-16'),color='r')
ax.set_title('Changes to Merrick Garland Wikipedia article, February 13 - March 18')
#ax.set_yscale('symlog')
ax.annotate('Official Announcement', xy=(pd.datetime(2016,3,16).date(),250), xytext=(pd.datetime(2016,3,4).date(), 300),
            arrowprops=dict(facecolor='black', shrink=0.05))

ax.annotate('???', xy=(pd.datetime(2016,3,14).date(),450), xytext=(pd.datetime(2016,3,11).date(), 400),
            arrowprops=dict(facecolor='black', shrink=0.05))

plt.savefig('merrick_garland_changes.png')
```

But let's compare Wikipedia's reaction to this event to Obama's prior announcements of Sonia Sotomayor and Elena Kagan.

```python
sonia_sotomayor_df = get_revision_df('Sonia Sotomayor')
elena_kagan_df = get_revision_df('Elena Kagan')
```

I'll assert what their nomination dates are to re-normalize time to days until/since nomination announcement.

```python
announcement_dates = {'Merrick Garland':pd.datetime(2016,3,16).date(),
                      'Elena Kagan':pd.datetime(2010,5,10).date(),
                      'Sonia Sotomayor':pd.datetime(2009,5,26).date()}
```

Compute time differences from annouoncement.

```python
merrick_garland_df['relative_date'] = (merrick_garland_df['date'] - announcement_dates['Merrick Garland'])/np.timedelta64(1,'D')
sonia_sotomayor_df['relative_date'] = (sonia_sotomayor_df['date'] - announcement_dates['Sonia Sotomayor'])/np.timedelta64(1,'D')
elena_kagan_df['relative_date'] = (elena_kagan_df['date'] - announcement_dates['Elena Kagan'])/np.timedelta64(1,'D')
```

Concatenate individual DataFrames together.

```python
obama_nominee_rev_df = pd.concat([merrick_garland_df,sonia_sotomayor_df,elena_kagan_df])
```

```python
agg_func = {'revid':{'number of revisions':len},
            'diff':{'median content difference':np.median,'cumulative content difference':np.sum},
            'latency':{'median latency':np.median}}
nominee_relative_activity = obama_nominee_rev_df.groupby(['page','relative_date']).aggregate(agg_func)
nominee_relative_activity.columns = nominee_relative_activity.columns.droplevel(0)
nominee_relative_activity.head()
```

```python
rev_activity_df = nominee_relative_activity['number of revisions'].unstack('page')
rev_activity_df = rev_activity_df.reindex_axis(np.arange(-730,730)).fillna(0)
ax = rev_activity_df.plot()
ax.set_xscale('symlog')
ax.set_xlabel('Time until/since nomination')
ax.set_ylabel('Revisions')
```

```python
rev_activity_df = nominee_relative_activity['median content difference'].unstack('page')
rev_activity_df = rev_activity_df.reindex_axis(np.arange(-730,730)).fillna(0)
ax = rev_activity_df.plot()
ax.set_xscale('symlog')
ax.set_ylim((-1e3,1e3))
ax.set_xlabel('Time until/since nomination')
ax.set_ylabel('Median bytes changed')
```

```python
rev_activity_df = nominee_relative_activity['cumulative content difference'].unstack('page')
rev_activity_df = rev_activity_df.reindex_axis(np.arange(-730,730)).fillna(0)
ax = rev_activity_df.plot()
ax.set_xscale('symlog')
ax.set_ylim((-2e3,2e4))
ax.set_xlabel('Time until/since nomination')
ax.set_ylabel('Cumulative bytes changed')
```

```python
rev_activity_df = nominee_relative_activity['median latency'].unstack('page')
rev_activity_df = rev_activity_df.reindex_axis(np.arange(-730,730)).fillna(method='bfill')
ax = rev_activity_df.plot()
ax.set_xscale('symlog')
ax.set_yscale('symlog')
ax.set_ylim((0,1e7))
ax.set_xlabel('Time until/since nomination')
ax.set_ylabel('Median latency')
```
